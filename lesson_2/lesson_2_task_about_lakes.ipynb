Generating data for the task

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
np.random.seed(2023)
data = np.random.randint(0, 100, 20)
data
```

```python
plt.plot(data)
```

searching for lowest points

```python
indices_min = np.where(data == np.min(data))[0]
indices_min
```

searching for deepest lake

```python
deepest_depth = 0
```

```python
for index in indices_min:
    left_bound = max(data[:index])
    right_bound = max(data[index + 1:])
    depth = min([right_bound, left_bound]) - data[index]
    if depth > deepest_depth:
        deepest_depth = depth
        deepest_lake = np.array([np.where(data == left_bound)[0][0], np.where(data == right_bound)[0][0], depth])

deepest_lake
```

retracting values for plotting the deepest lake

```python
deepest_lake_heights = [data[i] for i in range(deepest_lake[0], deepest_lake[1] + 1)]
deepest_lake_heights
```

plotting everything together

```python
plt.plot(data)
plt.plot(range(deepest_lake[0], deepest_lake[1] + 1), deepest_lake_heights, color='red')
```
